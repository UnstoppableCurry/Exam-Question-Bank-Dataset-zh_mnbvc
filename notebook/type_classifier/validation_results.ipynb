# 检查分类成试卷是否正确

```python
def query_file_total_by_keywords(df, keywords: list|set, uninclude_keywords: list|set = []):
    # keywords和uninclude_keywords可以被抽象理解成路径，如 "/题库"
    # keywords: 必须全部出现在"file_path"中，才会被统计
    # uninclude_keywords: 只要出现一个，就不会被统计

    # 创建一个全为 True 的布尔系列，它将用于存储是否每个文件路径包含所有关键字
    contains_all_keywords = pd.Series([True]*len(df))

    # 检查文件路径是否包含所有关键字
    for keyword in keywords:
        contains_all_keywords = contains_all_keywords & df['file_path'].fillna('').str.contains(keyword)

    if uninclude_keywords:
        # 检查文件路径是否包含 uninclude_keywords 中的任何一个
        for keyword in uninclude_keywords:
            contains_all_keywords = contains_all_keywords & ~df['file_path'].fillna('').str.contains(keyword)

    filtered_df = df.loc[contains_all_keywords]

    return filtered_df
```

# 在全部的范围中使用文件名推测

```python
def query_file_total_by_any_keywords(df, keywords: list|set, uninclude_keywords: list|set = []):
    # keywords和uninclude_keywords可以被抽象理解成路径，如 "/题库"
    # keywords: 只要一个出现在"file_path"中，就会被统计
    # uninclude_keywords: 只要出现一个，就不会被统计

    # 创建一个全为 False 的布尔系列，它将用于存储是否每个文件路径包含 keywords 中的任何一个
    contains_any_keywords = pd.Series([False]*len(df))

    # 检查文件路径是否包含 keywords 中的任何一个
    for keyword in keywords:
        contains_any_keywords = contains_any_keywords | df['file_path'].fillna('').str.contains(keyword)

    if uninclude_keywords:
        # 检查文件路径是否包含 uninclude_keywords 中的任何一个
        for keyword in uninclude_keywords:
            contains_any_keywords = contains_any_keywords & ~df['file_path'].fillna('').str.contains(keyword)
            
    filtered_df = df.loc[contains_any_keywords]

    return filtered_df
```

# 寻找假阴性

```python
keywords = {"真题", "试卷", "考试", "答案"}
uninclude_keywords = {
    "/千套简历模板+简历指导+面试真题-求职加薪必备", 
    "申论", 
    "/培训体系", 
    "书法教案", 
    "讲课内文言文精讲",
    "/复习知识点", 
    "简历",
    "调查表",
    "理解技巧",
    "考试大纲",
    "/知识星球", # 这个文件中所有文件看文件名都是试卷，但是实际不是试卷
}

# 全部文件中file_name包含关键字的文件
filtered_negative_file_df = query_file_total_by_any_keywords(negative_df, keywords, uninclude_keywords)
for _, row in filtered_negative_file_df.iterrows():
    print(row["file_path"])
filtered_negative_file_df
```
